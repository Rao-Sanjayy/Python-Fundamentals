### Importing necessary packages

```python
# !pip install pandasql
import sqlite3
import pandas as pd
from pandasql import sqldf
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#common fuction 
def pysqldf(q):
  return sqldf(q,globals())
```

### Batsmen Class

```python
class Batsmen:
  def __init__(self):
    pass
  
  def allBatters(self):
    try:
      out = pysqldf("SELECT * FROM batters;")
      out = out.set_index('Player')
      print(out)
      return out
    except:
      print("Some error has occured!")
    
  def topAvg5(self):
    try:
      out = pysqldf("select * from batters order by Average DESC limit 5;")
      out = out.set_index('Player')
      return print(out)
    except:
      print("Some error has occured!")
  
  def topAvg10(self):
    try:
      out = pysqldf("select * from batters order by Average DESC limit 10;")
      out = out.set_index('Player')
      return print(out)
    except:
      print("Some error has occured!")
  
  def topRun(self):
    out = pysqldf("select * from batters order by Runs DESC limit 5;")
    out = out.set_index('Player')
    return print(out)
  
  def bestStrikeR(self):
    out = pysqldf("select * from batters order by SR DESC limit 5;")
    out = out.set_index('Player')
    return print(out)

bat = Batsmen()
```

### Bowler Class

```python
class Bowler:
  def __init__(self):
    pass
  
  def allBowlers(self):
    try:
      out = pysqldf("SELECT * FROM bowlers;")
      out = out.set_index('Player')
      print(out)
      return out
    except:
      print("Some error has occured!")
    
  def topEco5(self):
    try:
      out = pysqldf("select * from bowlers where Overs > 100 order by Econ ASC limit 5;")
      out = out.set_index('Player')
      return print(out)
    except:
      print("Some error has occured!")
  
  def topEco10(self):
    try:
      out = pysqldf("select * from bowlers where Overs > 50 order by Econ ASC limit 10;")
      out = out.set_index('Player')
      return print(out)
    except:
      print("Some error has occured!")

  def topWickets(self):
    out = pysqldf("SELECT * FROM bowlers order by Wkts DESC limit 10;")
    out = out.set_index('Player')
    return print(out)

bowl = Bowler()
```

### Visualization Class

```python
class Visualization:
  def __init__(self):
    pass
  
  def plot_batter_avg(self):
    fig,ax = plt.subplots(figsize=(9,6))
    plt.xticks(rotation = 90)
    sns.barplot(x='Player',y='Average',data=batters,ci=95,ax=ax,
                order = batters.sort_values('Average').Player,palette = 'Spectral')
    ax.set_title('Overall Average of currently active players')
    plt.box(False)
    ax.tick_params(labelsize=12,length=0)
  
  def plot_bowler_eco(self):
    fig,ax = plt.subplots(figsize=(9,6))
    plt.xticks(rotation = 90)
    sns.barplot(x='Player',y='Econ',data=bowlers,ci=95,ax=ax,
                order = bowlers.sort_values('Econ').Player,palette = 'Spectral')
    ax.set_title('Overall Economy of currently active players')
    plt.box(False)
    ax.tick_params(labelsize=12,length=0)
  
draw = Visualization()
```

```python
playing_11 = set()
subs_4 = set() #empty set
a = bat.allBatters() 
b = bowl.allBowlers()
all_data = a.index.to_list() + b.index.to_list() #complete data in one object
print(all_data)
print(playing_11)
```

### Final selection method

```python
def final_11v3():
  try:
    global choice
    choice = input("Enter your selection: ")
    if choice in all_data:
      if choice not in playing_11:
        if len(playing_11)!=11:
          playing_11.add(choice)
          print("Player selected! \n")
        else:
          print("The team is full !")
      else:
        print("Player not found or already exists in the team.")
    else:
      raise ValueError
  except:
    print("Invalid input!")
```

```python
def sub_4():
  try:
    global choice
    choice = input("Enter your selection: ")
    if choice in all_data:
      if choice not in playing_11 or subs_4:
        if len(subs_4)!=4:
          subs_4.add(choice)
          print("Player selected! \n")
        else:
          print("The team is full !")
      else:
        print("Player not found or already exists in the team.")
    else:
      raise ValueError
  except:
    print("Invalid input!")
```
