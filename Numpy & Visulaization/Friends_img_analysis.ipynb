### Importing the necessary packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.image import imread
from scipy import linalg as lg
```

### Defining the image variable

```python
friends = imread("Forever_Friends.jpg")
print(type(friends))
```

```python
print(friends)
```

# Reshaping

```python
n = friends.size
N = 4
M = n//N
reshape_friends = friends.reshape((N,M))
print(reshape_friends) 
```

# Transpose

```python
friends_trans = friends.transpose()
print(friends_trans)
```

```python
n = friends_trans.size
N = 4
M = n//N
reshape_friends_trans = friends_trans.reshape((N,M))
print(reshape_friends_trans)
```

# Addition

```python
friends_add = reshape_friends + reshape_friends_trans
print(friends_add)
```

# Subtraction

```python
friends_subtract = reshape_friends - reshape_friends_trans
print(friends_subtract)
```

# Multiplication

```python
friends_multiply = reshape_friends * reshape_friends_trans
print(friends_multiply)
```

# Square Root

```python
np.sqrt(reshape_friends)
```

## Visualization

### 1. Line plot

```python
data = pd.read_csv('friends_dataset.csv - friends_dataset.csv.csv')

plt.figure(figsize = (12,8))

season_total = data[:24]["Episode Number"]
votes_1 = data[:24]["Votes"]
votes_2 = data[24:48]["Votes"]
votes_3 = data[48:72]["Votes"]

plt.style.use('fivethirtyeight')
plt.plot(season_total, votes_1, linestyle='--', linewidth = 2, marker = '.',markeredgecolor = 'black', label='Season 1 = Votes')
plt.plot(season_total, votes_2, linewidth = 2, marker = '.',markeredgecolor = 'black',label='Season 2 = Votes')
plt.plot(season_total, votes_3, linewidth = 2, marker = '.', markeredgecolor = 'black' ,label='Season 3 = Votes')

plt.xlabel('Episode Number', size = 12)
plt.ylabel('Votes', size = 12)
plt.title('Friends - Votes', size = 18)
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

### 2. Bar Graph

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize = (12, 8))

season_total = data[:24]["Episode Number"]
votes_4 = data[48:72]["Votes"]
votes_5 = data[72:96]["Votes"]
votes_6 = data[96:120]["Votes"]
           
x_indexes = np.arange(len(season_total))
width = 0.25
           
plt.bar(x_indexes - width, votes_4, width = width, label='Season 4 = Votes')
plt.bar(x_indexes, votes_5, width = width, label='Season 5 = Votes')         
plt.bar(x_indexes + width, votes_6, color='#467200', width = width, label='Season 6 = Votes')
plt.xticks(ticks = x_indexes, labels = season_total)
plt.xlabel('Episode Number')
plt.ylabel('Votes')
plt.title('Friends - Votes')
plt.legend()
plt.tight_layout()
plt.show()
```

### 4. Stack Plot

```python
plt.figure(figsize = (12, 8))
plt.style.use("fivethirtyeight")

season_total = data[:24]["Episode Number"]
votes_7 = data[120:144]["Votes"]
votes_8 = data[144:168]["Votes"]
votes_9 = data[168:192]["Votes"]
votes_10 = data[192:216]["Votes"]

labels = ['votes_7', 'votes_8', 'votes_9','votes_10']
colors = ['#6d904f', '#fc4f30', '#008fd5','#444444']

plt.stackplot(season_total, votes_7, votes_8, votes_9, votes_10, labels=labels, colors=colors)
plt.legend(loc=(0.07, 0.05))
plt.title("Friends - Vote")
plt.tight_layout()

plt.show()
```
